```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
movies = pd.read_csv('../data/csv/movies.csv')
```

```python
ratings = pd.read_csv('../data/csv/ratings.csv')
```

```python
tags = pd.read_csv('../data/csv/tags.csv')
```

```python
tags = tags.rename(columns = {"tag":"keywords"})
tags['keywords'].fillna('',inplace=True)
```

```python
merged = pd.merge(ratings, tags, how = 'left', left_on = ['userId','movieId'], right_on = ['userId','movieId'])
merged = pd.merge(merged, movies, how = 'left', on = 'movieId')
```

```python
merged ['genres'] = merged['genres'].str.replace("|"," ")
```

```python
merged['keywords'] = merged['keywords'] + " " + merged['genres']
merged['keywords'].fillna("",inplace=True)
```

```python
df = merged
```

```python
pivot_table = df.pivot_table(index = ["userId"],columns = ["title"],values = "rating")
pivot_table = pivot_table.fillna(0)
pivot_table
```

```python
users = df.userId.unique()
movies = df.movieId.unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
c_vect = TfidfVectorizer()
X = c_vect.fit_transform(merged['keywords'])
```

```python
cosine_sim = cosine_similarity(X)
```

```python
def get_movie_recommendation(movie_name):
    idx = merged[merged['title'].str.contains(movie_name)].index
    if len(idx):
        movie_indices = sorted(list(enumerate(cosine_sim[idx[0]])), key=lambda x: x[1], reverse=True)[1:6]
        movie_indices = [i[0] for i in movie_indices]
        return movie_indices
    else : 
        return []
```

```python
title = "Toy Story 2"
recommended_movie_list = get_movie_recommendation(title)
merged.iloc[recommended_movie_list].set_index('movieId')
```
